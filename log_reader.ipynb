```python
import json
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dqn_log_path = os.path.join(os.getcwd(),'dqn_implementation','elite_models_dqn','dqn-log')
evo_log_path = os.path.join(os.getcwd(),'elite_models_evo','acromuse-log')
```

```python
with open(dqn_log_path, 'r') as f:
    dqn_log = json.load(f)
```

```python
with open(evo_log_path, 'r') as f:
    evo_log = json.load(f)
```

```python
#extracting DQN log data
epoch = np.arange(92)+1
time = []
avg_score = []
max_score = []
for key in 50000*epoch:
    time.append(dqn_log[str(key)][0])
    avg_score.append(dqn_log[str(key)][2])
    max_score.append(dqn_log[str(key)][3])
time = np.array(time)/3600 # converting to hours
avg_score = np.array(avg_score)
max_score = np.array(max_score)
```

```python
#extracting evo log data
```

```python
plt.rcParams.update(
{
    'legend.fontsize':'x-large',
    'figure.figsize':(12,8),
    'axes.labelsize': 'x-large',
    'axes.titlesize': 'x-large',
    'xtick.labelsize':'x-large',
    'ytick.labelsize':'x-large'
})
```

```python
fig,ax = plt.subplots(1)
plt.title('DQN evaluation score each epoch',fontsize=20)
ax.plot(epoch,avg_score,label='Avg. score')
ax.plot(epoch,max_score,label='Max ep score')
ax.legend(loc=2)
```

```python
fig,ax = plt.subplots(1)
plt.title('Evaluation score per time',fontsize=20)
ax.plot(time,avg_score,label='Avg. score')
ax.plot(time,max_score,label='Max ep score')
ax.set_xlabel('Time [hours]')
ax.legend(loc=2)
```
